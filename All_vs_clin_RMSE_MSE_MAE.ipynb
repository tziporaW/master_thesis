```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import numpy as np
from google.colab import files
```

```python
metrics_path = "metrics_data.csv" # generate by merge_two_files
metric = pd.read_csv(metrics_path, dtype = 'object')
metrics_clin_path = "metrics_clin_data.csv" # generate by merge_two_files
metric_clin = pd.read_csv(metrics_clin_path, dtype = 'object')
```

```python
m=metric['Root Mean Square Error']
mc=metric_clin['Root Mean Square Error']
```

```python
m=pd.to_numeric(m, errors='coerce')
mc=pd.to_numeric(mc, errors='coerce')
```

```python
name=metric['Cancer Type']
```

```python
data = {'Name':name,
        'metrics':m, 
        'metric clinical':mc}
```

```python
df = pd.DataFrame(data)
```

```python
df_sort=df.sort_values('metrics')
```

```python
g_c=df_sort['metrics']
c=df_sort['metric clinical']
n=df_sort['Name']
```

```python
bar_plot=plt.figure()
metrics= g_c
metrics_clin = c
c_type = n
cat=['Genetics and Clinical', 'Clinical']

indices = np.arange(len(c_type))
width = 0.45
# Plotting
plt.bar(indices, g_c, width=width)
# Offsetting by width to shift the bars to the right
plt.bar(indices + width, c, width=width)

# Displaying year on top of indices
plt.xticks(ticks=indices, labels=c_type)
plt.xticks(rotation=90,fontsize=10)
plt.xlabel("Cancer Name",fontsize=16)
plt.ylabel("RMSE Value",fontsize=16)
plt.title("Importance of Gentic Feautures")
plt.legend(cat,loc=2)
plt.savefig('RMSE_Barplot.png')
plt.show()
files.download('RMSE_Barplot.png')
```
