```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import numpy as np
from google.colab import files
```

```python
metrics_path = "metrics_data.csv" # generate by merge_two_files
metric = pd.read_csv(metrics_path, dtype = 'object')
```

```python
rmse=metric['Root Mean Square Error']
mse=metric['Mean Square Error']
mae=metric['Mean Absolute Error']
r_2=metric['R Squared']
```

```python
rmse=pd.to_numeric(rmse, errors='coerce')
mse=pd.to_numeric(mse, errors='coerce')
mae=pd.to_numeric(mae, errors='coerce')
r_2=pd.to_numeric(r_2, errors='coerce')
```

```python
d = {'rmse': rmse, 'mse': mse,'mae': mae, 'r squared': r_2}
df = pd.DataFrame(data=d)
df
```

```python
name=metric['Cancer Type']
```

```python
import numpy as np; np.random.seed(0)
import seaborn as sns; sns.set_theme()

ax = sns.heatmap(df)
```

```python
bar_plot=plt.figure(figsize=(10, 8))

c_type = name
cat=['RMSE', 'MSE','MAE','R Squared']

indices = np.arange(len(c_type))
width = 0.25
# Plotting
plt.bar(indices, rmse, width=width)

# Offsetting by width to shift the bars to the right
plt.bar(indices + width, mse, width=width)
plt.bar(indices + (width*2), mae, width=width)
plt.bar(indices + (width*3), r_2, width=width)
# Displaying year on top of indices
plt.xticks(ticks=indices, labels=c_type)
plt.xticks(rotation=90,fontsize=15)
plt.xlabel("Cancer Name",fontsize=16)
plt.ylabel("Metric Value",fontsize=16)
plt.title("Metrics for Continuous Model",fontsize=16)
plt.legend(cat,loc=1)
plt.savefig('Cont_Barplot.pdf')
plt.show()
files.download('Cont_Barplot.pdf')
```
